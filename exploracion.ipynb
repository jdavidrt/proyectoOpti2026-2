# Exploración de datos — Optimización 2026-1

Análisis exploratorio de las instancias BPP y SCP.

```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

# Add models to path
sys.path.insert(0, os.path.join('..', 'models'))
from bin_packing import load_bin_packing
from set_covering import load_set_covering
```

## 1. Bin Packing — distribución de pesos

```python
n, C, weights = load_bin_packing('../data/BinPacking_u120_01.txt')
print(f'Items: {n}  |  Capacity: {C}')
print(f'Weight stats: min={min(weights)}, max={max(weights)}, mean={np.mean(weights):.1f}')

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(weights, bins=20, color='steelblue', edgecolor='white')
axes[0].set_title('Weight Distribution')
axes[0].set_xlabel('Weight'); axes[0].set_ylabel('Count')

# Theoretical lower bound: ceil(sum(w) / C)
lb = int(np.ceil(sum(weights) / C))
axes[1].bar(['Total weight', 'Capacity x LB'], [sum(weights), C * lb], color=['darkorange', 'steelblue'])
axes[1].set_title(f'Lower Bound = ceil({sum(weights)}/{C}) = {lb} bins')
plt.tight_layout(); plt.show()
```

## 2. Set Covering — densidad de la matriz de cobertura

```python
m, n, costs, coverage = load_set_covering('../data/SetCovering_50x200.txt')
print(f'Rows: {m}  |  Columns: {n}')
print(f'Cost stats: min={min(costs)}, max={max(costs)}, mean={np.mean(costs):.1f}')

# Build coverage matrix
A = np.zeros((m, n), dtype=int)
for i, cols in enumerate(coverage):
    for j in cols:
        A[i, j] = 1

density = A.sum() / (m * n)
print(f'Matrix density: {density:.2%}')

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].imshow(A, aspect='auto', cmap='Blues')
axes[0].set_title('Coverage Matrix A (blue = covered)')
axes[0].set_xlabel('Column'); axes[0].set_ylabel('Row')

axes[1].hist(costs, bins=20, color='firebrick', edgecolor='white')
axes[1].set_title('Column Cost Distribution')
axes[1].set_xlabel('Cost'); axes[1].set_ylabel('Count')
plt.tight_layout(); plt.show()
```
